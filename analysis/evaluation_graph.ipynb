```python
import matplotlib.pyplot as plt

# Base tier costs (e.g. 1024 MB)
cost = [1.25, 72.50, 786.80]
num_reqs = [100_000, 1_000_000, 10_000_000]

# 2048 MB tier costs
cost_2048 = [9.17, 151.67, 1578.47]
num_reqs_2048 = [100_000, 1_000_000, 10_000_000]

# Create the plot
fig, ax = plt.subplots()

# Plot both lines
ax.plot(num_reqs, cost, 'o-', label='1024 MB', color='teal')
ax.plot(num_reqs_2048, cost_2048, 's--', label='2048 MB', color='orange')

# Annotate data points
for x, y in zip(num_reqs, cost):
    ax.text(x, y, f"${y:.2f}", fontsize=8, ha='left', va='bottom')
for x, y in zip(num_reqs_2048, cost_2048):
    ax.text(x, y, f"${y:.2f}", fontsize=8, ha='left', va='bottom')

# Use log scale for X and Y
ax.set_xscale('log')


# Labels and title
ax.set_xlabel("Requests per Month", fontsize=10)
ax.set_ylabel("Cost per Month (USD)", fontsize=10)
ax.set_title("Monthly Cost vs Requests with Different Memory Sizes")

# Grid and legend
ax.legend()

# Format X-axis with commas
ax.set_xticks(num_reqs)  # Set custom ticks if desired
ax.set_xticklabels([f"{x:,}" for x in num_reqs])  # Adds commas

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Requests per Month": [100_000, 1_000_000, 10_000_000],
    "Cost @ 1024 MB (USD)": [1.25, 72.50, 786.80],
    "Cost @ 2048 MB (USD)": [9.17, 151.67, 1578.47]
}

# Format data using pandas
df = pd.DataFrame(data)
df["Requests per Month"] = df["Requests per Month"].apply(lambda x: f"{x:,}")
df["Cost @ 1024 MB (USD)"] = df["Cost @ 1024 MB (USD)"].apply(lambda x: f"${x:,.2f}")
df["Cost @ 2048 MB (USD)"] = df["Cost @ 2048 MB (USD)"].apply(lambda x: f"${x:,.2f}")

# Create a Matplotlib figure and table
fig, ax = plt.subplots(figsize=(8, 2))
ax.axis('off')  # Hide the axes

# Add the table
table = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc='center',
    loc='center'
)

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)  # Widen cells for readability

plt.title("Monthly Cost Table", fontsize=12, pad=20)
plt.tight_layout()
plt.show()
```
